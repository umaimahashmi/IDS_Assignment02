**`PART: e`**

---
Identify if there are any outliers in your datasets (for all datasets
2015-2019)



```python
#identifying the outliers in the data set of 2015
z_score_threshold = 3
outliers = pd.DataFrame()

for column in df_2015.select_dtypes(include=[np.number]):
    z_scores = np.abs((df_2015[column] - df_2015[column].mean()) / df_2015[column].std())
    column_outliers = z_scores > z_score_threshold
    outliers[column] = column_outliers

row_outliers = outliers.any(axis=1)
outliers_data = df_2015[row_outliers]
#The "outliers_data" DataFrame is created by selecting rows from df_2015 based on
#the row_outliers boolean values. It contains the rows with at least one outlier in any of the numeric columns.

outliers_data
df_2015_no_outliers = df_2015[~row_outliers]
df_2015_no_outliers    #contains rows with outliers removed
```

```python
#identifying the outliers in the data set of 2015
z_score_threshold = 3
outliers = pd.DataFrame()

for column in df_2016.select_dtypes(include=[np.number]):
    z_scores = np.abs((df_2016[column] - df_2016[column].mean()) / df_2016[column].std())
    column_outliers = z_scores > z_score_threshold
    outliers[column] = column_outliers

row_outliers = outliers.any(axis=1)
outliers_data = df_2016[row_outliers]
#The "outliers_data" DataFrame is created by selecting rows from df_2015 based on
#the row_outliers boolean values. It contains the rows with at least one outlier in any of the numeric columns.

outliers_data
df_2016_no_outliers = df_2016[~row_outliers]
df_2016_no_outliers.shape #contains rows with outliers removed
```

```python
import matplotlib.pyplot as plt

# Select the "Happiness Score" column from the "outliers_data" DataFrame
happiness_scores = outliers_data['Happiness Score']

# Create a scatter plot to visualize the outliers in the "Happiness Score" column
plt.figure(figsize=(10, 6))
plt.scatter(range(len(happiness_scores)), happiness_scores, color='blue', label='Data Points')
plt.scatter(happiness_scores[happiness_scores > 3].index, happiness_scores[happiness_scores > 3], color='red', label='Outliers')
plt.xlabel('Index')
plt.ylabel('Happiness Score')
plt.title('Scatter Plot of Happiness Score with Outliers Highlighted')
plt.legend()
plt.show()
```

```python
# Create a figure and axes for the box plots
fig, ax = plt.subplots(figsize=(12, 6))

# Select relevant columns for the box plot
selected_columns = df_2015[['Region', 'Happiness Score']]

# Create box plots of Happiness Score by Region, highlighting outliers
sns.boxplot(x='Region', y='Happiness Score', data=selected_columns, ax=ax, flierprops=dict(markerfacecolor='red', markersize=4))

# Highlight the outliers using scatter plot
sns.scatterplot(data=outliers_data, x='Region', y='Happiness Score', ax=ax, color='red', label='Outliers')

# Set labels and title
ax.set_xlabel("Region")
ax.set_ylabel("Happiness Score")
ax.set_title("Box Plots of Happiness Score by Region with Outliers Highlighted")

# Show the legend for the outliers
ax.legend()

# Customize x-axis label rotation and font size
ax.tick_params(axis='x', labelrotation=45, labelsize=10)

# Show the box plots with outliers highlighted
plt.show()
```

---
Identify which 2 variables have the highest correlations across all the countries for the years 2015 and 2016.


```python
#part j Identify which 2 variables have the highest correlations across all the
#countries for the years 2015 and 2016.

df_2015_numeric=df_2015_no_outliers.dropna(subset=['Country', 'Region'])
correlation_matrix = df_2015_numeric.corr() # Calculate the correlation matrix

# Find the pair of variables with the highest correlation
max_corr = 0
variable1, variable2 = None, None

# Iterate through the correlation matrix
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > max_corr:
            max_corr = abs(correlation_matrix.iloc[i, j])
            variable1 = correlation_matrix.columns[i]
            variable2 = correlation_matrix.columns[j]

print(f"The pair of variables with the highest correlation is '{variable1}' and '{variable2}' with a correlation coefficient of {max_corr:.2f}")
```
